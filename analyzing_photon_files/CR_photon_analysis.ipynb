```python
x_coor = set()
y_coor = set()
x_cosine = set()
y_cosine = set()
height = set()
time = set()
wavelength = set()
particle_type = set()
tel_id = set()
```

```python
wavelength_list = []
```

```python
for data_file in (files):
    with open(data_file, 'r') as f:

        for line in tqdm(f, 'Parsing datafile'):
            if line.startswith('P '):
                components = line.split()

                x_coor.add(components[1])
                y_coor.add(components[2])
                x_cosine.add(components[3])
                y_cosine.add(components[4])
                height.add(components[5])
                time.add(components[6])
                wavelength.add(components[7])
                particle_type.add(components[8])
                tel_id.add(components[9])

                wavelength_list.append(int(components[7]))
```

```python
from astropy import units as u
from astropy import constants as c

P_std = 101325
T_15C = 288.15
e_charge = 1.602e-19
mu_air = 1.00000037*c.mu0.value

def n_air(lam, P=P_std, T=T_15C):
    
    return 1 + 7.76e-7 * (1 + 7.52e-15 / lam**2) * P/T

def d2Ndxdlam(lam, beta, q=e_charge, mu=mu_air):

    return np.pi * q**2 * mu / c.h.value * (c.c.value / lam**2) * (1 - 1 / (beta * n_air(lam))**2)
```

```python
def get_beta(E, m=c.m_e.value):

    return np.sqrt(1 - m**2 * c.c.value**4 / E**2)
```

```python
muon_mass = (105.65837 * u.MeV/c.c**2).to(u.kg).value
muon_mass
```

```python
beta = get_beta(E=1 * u.TeV.to(u.J), m=muon_mass)
```

```python
atmos_height = 112.8e3  # m
```

```python
wvl_arr = np.linspace(250, 700, 1000)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

plt.hist(wavelength_list, bins=histobins, density=False, color='black')
coeff = atmos_height * d2Ndxdlam(300e-9, beta) / (45000*binsize)
plt.plot(wvl_arr, atmos_height * d2Ndxdlam(wvl_arr * 1e-9, beta) / coeff, c='firebrick', lw=2, 
         label='Frank-Tamm')

plt.grid(alpha=0.8)
plt.ylabel(r'$\frac{dN}{d\lambda}$  [nm$^{-1}$]')
plt.xlabel('Cherenkov photon wavelength  [nm]')
plt.title('Cherenkov spectrum histogram')

ticks = ax.yaxis.axes.get_yticks()
ticks_adjust = [int(tick / binsize) for tick in ticks]
plt.yticks(ticks=ticks, labels=ticks_adjust)

plt.legend()
plt.show()
```
