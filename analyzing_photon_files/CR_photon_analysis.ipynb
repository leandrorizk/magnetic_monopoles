```python
wavelength_list = []
```

```python
for data_file in (files):
    with open(indir+data_file, 'r') as f:

        for line in tqdm(f, 'Parsing datafile'):
            if line.startswith('P '):
                components = line.split()

                # x_coor.add(components[1])
                # y_coor.add(components[2])
                # x_cosine.add(components[3])
                # y_cosine.add(components[4])
                # height.add(components[5])
                # time.add(components[6])
                wavelength_list.append(int(components[7]))
                particle_type.add(components[8])
                # tel_id.add(components[9])

wavelength = set(wavelength_list)
```

```python
binsize = 10
histobins = np.arange(250, 710, binsize)
histobins
```

```python
wavelength_array = np.array(wavelength_list)
freq_array = 3e8 / (wavelength_array * 1e-9)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

plt.hist(wavelength_list, bins=histobins, density=False, color='black')

plt.grid(alpha=0.8)
plt.ylabel(r'$\frac{dN}{d\lambda}$  [nm$^{-1}$]')
plt.xlabel('Cherenkov photon wavelength  [nm]')
plt.title('Cherenkov spectrum histogram from hadronic showers')

ticks = ax.yaxis.axes.get_yticks()
ticks_adjust = [int(tick / binsize) for tick in ticks]
plt.yticks(ticks=ticks, labels=ticks_adjust)

plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.hist(freq_array, bins=1000, density=False, color='black')

plt.grid(alpha=0.8)
plt.ylabel('Occurence')
plt.xlabel('Cherenkov photon frequency  [Hz]')
plt.title('Cherenkov spectrum histogram from hadronic showers')

plt.show()
```
